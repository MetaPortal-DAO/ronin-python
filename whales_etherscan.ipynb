```python
import pandas as pd 
from dotenv import load_dotenv
import os
import requests
import json
load_dotenv()


# forgot that you can't just hit mainnet!
marketplace_whales = pd.read_json("./data/whales_by_value_transfer.json")
```

```python
top_10_market_whales = marketplace_whales.sort_values("value", ascending=False).reset_index()[0:10].rename(columns = {"index": "address"})

market_whale_dct = {}

for index, arr in top_10_market_whales.iterrows():
    addy = "0x" + arr[0]
    url = f'https://api.covalenthq.com/v1/1/address/{addy}/transactions_v2/?quote-currency=USD&format=JSON&block-signed-at-asc=false&no-logs=true&key={os.getenv("COVALENT_API_KEY")}'
    res = requests.get(url)
    insrt = json.loads(res.text)['data']['items']
    market_whale_dct[addy] = insrt
```

```python
market_whales_total = []
for index, arr in top_10_market_whales.iterrows():
    addy = "0x" + arr[0]
    market_whale_dct_summary = {}
    for item in market_whale_dct[addy]:
        print(item)
        datadump = []
        datadump.extend([item['block_signed_at'], item["to_address"], item["value_quote"]])
        market_whale_dct_summary[addy] = datadump
    market_whales_total.append(market_whale_dct_summary)
```
